# 5. Visualization

## 1.Histogram

```python
plt.figure(figsize=(7,7))
plt.hist(iris['SL'],color='red',rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(7,7))
plt.hist(iris['SW'],color='blue',rwidth=0.9)
plt.title('Distribution of SW',fontsize=16)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(7,7))
plt.hist(iris['PL'],color='red',rwidth=0.9)
plt.title('Distribution of PL',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(7,7))
plt.hist(iris['PW'],color='blue',rwidth=0.9)
plt.title('Distribution of PW',fontsize=16)
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.hist(iris['PL'])
plt.hist(iris['PW'])
```

## 2. Scatter Plot


```python
plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.scatter(iris['PL'],iris['PW'],s=10,c='red',marker='^')
plt.title('Petal Length vs Petal Width', fontsize=14)
plt.xlabel('Petal Length',fontsize=10)
plt.ylabel('Petal Width',fontsize=10)
           
plt.subplot(2,1,2)
plt.scatter(iris['SL'],iris['SW'],s=10,c='blue',marker='^')
plt.title('Sepal Length vs Sepal Width', fontsize=14)
plt.xlabel('Sepal Length',fontsize=10)
plt.ylabel('Sepal Width',fontsize=10)

plt.tight_layout()
```
